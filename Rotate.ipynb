# Rotate Images

```python
import numpy
import cv2
from matplotlib import pyplot as plt
%matplotlib inline
```

# Bild laden

```python
file = 'documents/skewed.tiff'
im = cv2.imread(file, 0)
plt.figure(figsize=(16, 16))
plt.imshow(im)
```

# Thresholding

```python
_, binary_thresh = cv2.threshold(im, 200, 255, cv2.THRESH_BINARY_INV)

fig = plt.figure(figsize=(12, 12))
fig.add_subplot(1, 2, 1)
plt.imshow(im)

fig.add_subplot(1, 2, 2)
plt.imshow(binary_thresh)
```

# Detect lines

```python
vis = cv2.cvtColor(im, cv2.COLOR_GRAY2RGB)
orig = vis.copy()

lines = cv2.HoughLinesP(binary_thresh, 1, numpy.pi/180, 100, minLineLength= 600/2.0, maxLineGap=20)
angle = 0
for line in lines:
    x1, y1, x2, y2 = line[0]
    r = numpy.arctan2(y2 - y1, x2 - x1)
    cv2.line(pt1=(x1,y1),pt2=(x2,y2),img=vis,color=(255,0,255),thickness=1)
    angle += numpy.arctan2(y2 - y1, x2 - x1)

avg_radian = angle / len(lines)
avg_angle = avg_radian * 180 / numpy.pi
print "Average angle is %f degrees" % avg_angle

(h, w) = im.shape
center = (w // 2, h // 2)
matrix = cv2.getRotationMatrix2D(center, avg_angle, 1.0)
rotated = cv2.warpAffine(im,
                         matrix,
                         (w, h),
                         flags=cv2.INTER_CUBIC,
                         borderMode=cv2.BORDER_REPLICATE)

fig = plt.figure(figsize=(20, 20))
fig.add_subplot(1, 3, 1)
plt.imshow(orig)

fig.add_subplot(1, 3, 2)
plt.imshow(vis)

fig.add_subplot(1, 3, 3)
plt.imshow(rotated)
```
